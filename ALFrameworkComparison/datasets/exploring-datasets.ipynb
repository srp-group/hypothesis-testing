```python
import os
import numpy as np
```

### DNA

```python
current_root = os.getcwd()
data_path = os.path.join(current_root, "..", "..", "data", "dna", "dna.npz")
data_path = os.path.normpath(data_path)
with np.load(data_path, allow_pickle=True) as file:
    x = file["x"].astype(np.float32)
    y = file["y"].astype(np.float32)
```

```python
print(x.shape, y.shape)
```

```python
cov_matrix = np.cov(np.concatenate((x, y), axis=1), rowvar=False)
print(cov_matrix.shape)
```

```python
class_counts = np.sum(y, axis=0)
print(class_counts)
```

```python
corelated_features = cov_matrix[:,-3:].argmax(axis=1)
classes, counts = np.unique(corelated_features, return_counts=True)
classes, counts
```

```python
from sklearn.decomposition import PCA

# Assuming the sparse features are stored in a variable called 'sparse_features'
# sparse_features.shape should be (n_samples, 180)

# Create an instance of PCA with 3 components
pca = PCA(n_components=3)

# Fit the PCA model to the sparse features and transform them to the latent space
latent_embedding = pca.fit_transform(x)

# The latent_embedding variable now contains the 3-dimensional embedding
latent_embedding.shape
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Assuming 'x' is your data and 'y' is your labels

# Convert one-hot encoded labels to single numbers for color coding
y_color = np.argmax(y, axis=1)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(latent_embedding[:, 0], latent_embedding[:, 1], latent_embedding[:, 2], c=y_color, cmap='viridis')

# Create a legend for the color coding
legend1 = ax.legend(*scatter.legend_elements(), title="Classes")
ax.add_artist(legend1)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

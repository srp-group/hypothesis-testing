```python
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
instances = 20
k_repetitions = 10
k_range = np.linspace(0, k_repetitions - 1, num=k_repetitions)

regularizations = ['L2', 'dropout']
l2_no = 3
l2_range = (0.001, 0.1)
drop_no = 3
drop_range = (0.05, 0.25)

l2_values = np.linspace(l2_range[0], l2_range[1], l2_no)
drop_values = np.linspace(drop_range[0], drop_range[1], drop_no)

print(l2_values)
print(drop_values)

# Define parameters for Gaussian noise
mean = 0  # Mean of the Gaussian distribution
std_dev = 0.1  # Standard deviation of the Gaussian distribution
```

```python
def calculate_x(x_value):
    return 0.1 * np.exp(-1 * x_value)

def calculate_y(y_value):
    return 0.1 * y_value ** 2

def create_sequence(sequence_list, df):
    df_length = df.shape[0]
    # repeated_sequence = sequence_list * (df_length // len(sequence_list)) + sequence_list[:df_length % len(sequence_list)]
    repeated_sequence =  [sequence_list[i % len(sequence_list)] for i in range(df_length)]
    return repeated_sequence

def sequence(sequence_list, df):
    df_length = df.shape[0]
    # sequence = sequence_list[0] * (df_length // 2) + sequence_list[1] * (df_length // 2)
    first_half_repeats = df_length // 2
    second_half_repeats = df_length - first_half_repeats

    # Create a column with alternating items from the sequence list
    sequence = [sequence_list[0]] * first_half_repeats + [sequence_list[1]] * second_half_repeats
    return sequence
```

```python
df = pd.DataFrame()#columns = ['added_instances'])#, 'repetition', 'reg_type', 'reg_value'])
```

```python
df['id'] = range(instances * len(regularizations) * (l2_no + drop_no) * k_repetitions)
print(df.shape)
```

```python
instances_list = np.linspace(0, instances-1, instances)
instance_reps = len(regularizations) * (l2_no + drop_no) * k_repetitions

instance_sequence = create_sequence([x for x in instances_list for _ in range(len(instances_list))], df)
df['instance_no'] = instance_sequence
print(df.shape)
print(df.head(60))
print(df.tail(60))
# print(df.iloc[80:120,:])
```

```python
reg_sequence = sequence(regularizations, df)
df['reg_type'] = reg_sequence
print(df.shape)
print(df.head(20))
print(df.tail(20))
```

```python
k_sequence = create_sequence(k_range, df)
# len(k_sequence)
df['k_repetition'] = k_sequence
print(df.shape)
print(df.head(20))
print(df.tail(20))
```

```python
l2_sequence = create_sequence([x for x in l2_values for _ in range(k_repetitions)], df) 
drop_sequence = create_sequence([x for x in drop_values for _ in range(k_repetitions)], df)

sequences = {regularizations[0]: l2_sequence, regularizations[1]: drop_sequence}

df['reg_value'] =  [sequences[row['reg_type']][i % len(sequences[row['reg_type']])] for i, row in df.iterrows()]
print(df.shape)
print(df.head(40))
print(df.tail(40))
```

```python
df['loss'] = (np.exp(-1 * df['instance_no'])) + (10 * df['reg_value'] ** 2)
# df['loss'] = calculate_x(df['instance_no']) + calculate_y(df['reg_value'])
df['loss'] = df['loss'] + np.random.normal(mean, std_dev, len(df))
df.head(20)
```

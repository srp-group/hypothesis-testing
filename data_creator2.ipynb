```python
import pandas as pd
import random
import plotly.express as px
import numpy as np
```

```python
seed_no = 1

random.seed(seed_no)
np.random.seed(seed_no)

instances = 100
regularizations = ['L2', 'dropout']
l2_range = (0.0001, 0.01)
drop_range = (0.1, 0.5)

# Define parameters for Gaussian noise
mean = 0  # Mean of the Gaussian distribution
std_dev = 0.01  # Standard deviation of the Gaussian distribution
```

```python
df = pd.DataFrame()
df['instance_no'] = range(instances)
df['L2_value'] = [random.uniform(l2_range[0], l2_range[1]) for _ in range(instances)]
df['drop_value'] = [random.uniform(drop_range[0], drop_range[1]) for _ in range(instances)]

df['loss'] = (np.exp(-0.1 * df['instance_no'])) + (0.01 * df['L2_value'] ** 2 + 0.01 * df['drop_value'] ** 2)
df['loss'] = df['loss'] + np.random.normal(mean, std_dev, len(df)) #Adding noise to the toy dataset

print(df.shape)
df.head(5)
```

```python
df.to_csv('ToyData3.csv', index=False)
```
